The Ames data set contains of 81 columns. This is way to many to make sense of. Let's try to identify which of these attributes have the biggest impact on the house `Sales Price`.

# Null values
First let's study what columns have lot's of null values.

```python
df = data.isnull().sum().sort_values(ascending=False)
df[df>0]
```

Some features like PoolQC, MiscFeature, Alley and Fence are missing for over 1000 households (1460 in the dataset). The `data_description.txt` file tells us that this means a lock of pool etc, but still we can't make much of an attribute that is the same for most households.

# Data types

```python
data.dtypes.value_counts()
```

We have 38 numerical values, and 43 string types. However, without looking at the values it is hard to tell if the variables are categorical or quantitative. A numerical could take limited values and represent a categorical, and a string could represent a date etc.

```python
unique = pd.DataFrame(data.nunique().sort_values(ascending=False)).reset_index().rename(columns={0 : 'unique'})
dtypes =  pd.DataFrame(data.dtypes).reset_index().rename(columns={0 : 'dtype'})
entries =  pd.DataFrame(data.notnull().sum()).reset_index().rename(columns={0 : 'entries'})
```

```python
unique.head(1)
```

```python
df = pd.merge(unique, entries, on = 'index')
df = pd.merge(df, dtypes, on = 'index')
```
